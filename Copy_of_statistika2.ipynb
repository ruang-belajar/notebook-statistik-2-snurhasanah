```python
# PERSIAPAN DATA

# download file contoh csv
!wget https://raw.githubusercontent.com/ruang-belajar/python/master/notebook/data/2013-data-kebakaran.csv -O data-kebakaran.csv


import pandas as pd

# load data ke dataframe (df)
df = pd.read_csv("data-kebakaran.csv")
df.head()
```

1. Filter kejadian kebakaran yang terjadi di bulan Maret

```python
df_march = df[df['bulan'] == 3]
print(df_march)
```

2. Filter kejadian kebakaran yang nilai kerugiannya antara 50jt sampai 100jt


```python
df_loss = df[(df['kerugian'] >= 50000000) & (df['kerugian'] <= 100000000)]
print(df_loss)
```

3. Tampilkan/Filter 5 kejadian yang tingkat kerugiannya paling besar


```python
df_top5 = df.nlargest(5,'kerugian')
print(df_top5)
```

4. Menggunakan groupby dan count, hitung banyaknya kejadian kebakaran per bulan


```python
df_count = df.groupby('bulan').size().reset_index(name='count')
print(df_count)
```

5. Menggunakan groupby dan sum, hitung jumlah kerugian akibat kebakaran per bulan


```python
df_sum = df.groupby('bulan')['kerugian'].sum().reset_index(name='total_kerugian')
print(df_sum)
```

6. Menggunakan gruppby dan mean, hitung rata-rata kerugian akibat kebakaran per bulan


```python
df_mean = df.groupby('bulan')['kerugian'].mean().reset_index(name='mean_kerugian')
print(df_mean)
```

7. Hitung banyaknya kejadian kebakaran yang disebabkan Listrik per bulan


```python
df_electric = df[df['penyebab'] == 'listrik']
df_electric_count = df_electric.groupby('bulan').size().reset_index(name='count')
df

```

8. Hitung jumlah kerugian akibat kebakaran yang terjadi di kelurahan Kemayoran


```python
import pandas as pd

kemayoran_fires = df[df['kelurahan'] == 'Kemayoran']

total_kerugian_kemayoran = kemayoran_fires['kerugian'].sum()

print("Total Kerugian in Kemayoran:", total_kerugian_kemayoran)
```

9. Hitung rata-rata kerugian akibat kebakaran yang terjadi di bulan Juni dan Juli,  di Pasar Rebo


```python
df_june_july_rebo = df[(df['bulan'] == 'juni') | (df['bulan'] == 'juli') & (df['kelurahan'] == 'PASAR REBO')]
mean_loss_june_july_rebo = df_june_july_rebo['kerugian'].mean()
print(mean_loss_june_july_rebo)
```

10. Menggunakan fungsi max, tampilkan nilai kerugian paling besar dari kejadian kebakaran per kelurahan.

```python
df_max_loss_kelurahan = df.groupby('kelurahan')['kerugian'].max().reset_index(name='max_kerugian')
print(df_max_loss_kelurahan)
```
